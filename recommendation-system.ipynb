```python
import kagglehub
import numpy as np # linear algebra
import pandas as pd
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
path =kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")


print(path)
```

```python
import pandas as pd 
books = pd.read_csv(f"{path}/books.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax = plt.axes()

sns.heatmap(books.isna().transpose(),cbar=False,ax=ax)

plt.xlabel("Column value")
plt.ylabel("Missing Column")
plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books['description'].isna(),0,1)

books["age_of_book"] =  2025 - books["published_year"]


```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]
```

```python

# multiple column nite HOLE double bracket use korte hobe
correlation_matrix = books[columns_of_interest].corr(method="spearman")

# plot style + figure
sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

# heatmap draw
heatmap = sns.heatmap(
    correlation_matrix,
    annot=True,          # cell er moddhe value show korbe
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1, vmax=1,
    linewidths=0.5
)

# title set kora
heatmap.set_title("Correlation Heatmap (Spearman)", pad=12)

plt.tight_layout()
plt.show()
```

```python
books[(
    books["description"].isna()
    |books["num_pages"].isna()
    |books["average_rating"].isna()
    |books["published_year"].isna()
)]
```

```python
book_missing=books[(
    ~books["description"].isna()
    & ~books["num_pages"].isna()
    & ~books["average_rating"].isna()
    & ~books["published_year"].isna()
)]



```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing = book_missing.copy()

book_missing['word_of_description'] = (
    book_missing['description'].str.split().str.len()
)
```

```python
book_missing
```

```python
book_missing.loc[book_missing['word_of_description'].between(1,4),'description']
```

```python
book_missing.loc[book_missing['word_of_description'].between(5,9),'description']
```

```python
book_missing_25_words = book_missing[book_missing['word_of_description']>=25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words= book_missing_25_words.copy()

book_missing_25_words["title and subtitle"] = np.where(
    book_missing_25_words['subtitle'].isna(),
    book_missing_25_words['title'],
    book_missing_25_words[['title', 'subtitle']].astype(str).agg(":".join, axis=1)
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words['tagged_description'] = (
    book_missing_25_words[['isbn13', 'description']]
    .astype(str)
    .agg("".join, axis=1)
)
```

```python
book_missing_25_words
```

```python
(book_missing_25_words.drop(['subtitle','missing_description','age_of_book','word_of_description'],axis=1)).to_csv("book_cleaned_csv",index=False)
```
